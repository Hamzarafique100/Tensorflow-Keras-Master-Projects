```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from PIL import Image, ImageTk
import tensorflow as tf
import numpy 

from tensorflow.keras.models import load_model
model = load_model('model1_image recognition_50epoch.h5')
classes = { 
    0:'aeroplane',
    1:'automobile',
    2:'bird',
    3:'cat',
    4:'deer',
    5:'dog',
    6:'frog',
    7:'horse',
    8:'ship',
    9:'truck' }

class Root(Tk):
    def __init__(self):
        super(Root, self).__init__()
        self.title("Image classification with CNN")
        self.minsize(750, 500)
        self.configure(background='#CDCDCD')

        self.labelFrame = ttk.LabelFrame(self, text = "Open File")
        self.labelFrame.grid(column = 0, row = 1, padx = 20, pady = 20)

        self.buttons()
    def classify(self):
        global label_packed
        image = Image.open(self.filename)
        image = image.resize((32,32))
        image = numpy.expand_dims(image, axis=0)
        image = numpy.array(image)
        y_prob = model.predict([image])[0]
        pred = y_prob.argmax(axis=-1)

        sign = classes[pred]
        print(sign)
        
        self.label3=Label(foreground='#011638', text=sign,font=('arial',15,'bold')) 
        self.label3.grid(column=3,row=15)


    def buttons(self):
        self.button1 = ttk.Button(self.labelFrame, text = "Upload an image",command = self.fileDialog)
        self.button1.grid(column = 1, row = 1)
        self.button2 = ttk.Button(master=self, text="Classify image",command=self.classify)
        self.button2.grid(column = 0, row = 9)
        
    def fileDialog(self):

        self.filename = filedialog.askopenfilename(initialdir =  "C:\\Users\\Hamza Rafique\\Downloads\\", title = "Select A File", filetype =
        (("jpeg files","*.jpg"),("all files","*.*")) )
        self.label = ttk.Label(self.labelFrame, text = "")
        self.label.grid(column = 1, row = 2)
        self.label.configure(text = self.filename)

        img = Image.open(self.filename)
        photo = ImageTk.PhotoImage(img)

        self.label2 = Label(image=photo)
        self.label2.image = photo
        self.label2.grid(column=3, row=3)

root = Root()
root.mainloop()
```
